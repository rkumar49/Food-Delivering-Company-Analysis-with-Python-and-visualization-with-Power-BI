```python
# Zamoto Analysis

# Import Libaries
import pandas as pd
import numpy as np

# Loading .csv file of zomato data using pandas
df_data = pd.read_csv(r"C:\Users\Rahul\Downloads\Zomato Data Analaysis\uber eats.csv", encoding='unicode_escape')
print(df_data)
```

```python


# Loading excel file of country code using pandas

df_country = pd.read_excel(r"C:\Users\Rahul\Downloads\Zomato Data Analaysis\Country-Code.xlsx")
print(df_country)

```

```python

# Data Cleaning (Zamaoto data)
#check the datatype
df_data.info()

# Check the Null Values
print(df_data.isnull().sum())
```

```python

# the top 5 rows of sales data
print(df_data.head(10))

# describe return description of the data in the DataFrame
print(df_data.describe())

# Check Duplicates
print(df_data.duplicated().sum())

# Columns Names
print(df_data.columns)

#check the datatype
print(df_data.dtypes)

print(df_data.shape)


# now clean the Country Code Data

# view data of country code
print(df_country.head())

print(df_country.info)

print(df_country.shape)
```

```python

#checking null values
print(df_country.isnull().sum())

# checking Duplicates
df_country.duplicated().sum()


# Merging country code and Data file)
df=pd.merge(df_data,df_country,on='Country Code', how= 'left')
print(df.head())


#droping unwanted columns
df.drop(['Restaurant ID','Locality Verbose','Country Code','Longitude','Latitude'],axis=1,inplace=True)
print(df.columns)

# checking the null values after merging
df.isnull().sum()

# remove null values
df=df.dropna()
print(df.isnull().sum())

# count total transcation happen in all the world
print(df.Country.value_counts())


#downlaod data
df.to_csv(r"C:\Users\Rahul\Downloads\Zomato Data Analaysis\Clean Uber eats Data.csv")
print(df.to_string)
```

```python
total_restaurants = df['Restaurant Name'].nunique()
total_cuisines = df['Cuisines'].nunique()
print(f"Total Restaurants: {total_restaurants}, Total Cuisines: {total_cuisines}")
```

```python
restaurants_by_country = df['Country'].value_counts()
print("Restaurants by Country:\n", restaurants_by_country)
```

```python
india_df = df[df['Country'] == 'India']
value_for_money_restaurants = india_df[india_df['Rating text'].str.contains("Good|Very Good|Excellent", case=False)]
cities_with_value_restaurants = value_for_money_restaurants['City'].value_counts()
print("Indian Cities with the Most 'Value for Money' Restaurants:\n", cities_with_value_restaurants)
```

```python
top_cuisines_india = india_df.groupby('Cuisines')['Votes'].sum().nlargest(10)
print("Top 10 Cuisines with Most Votes in India:\n", top_cuisines_india)
```

```python
online_delivery_countries = df[df['Has Online delivery'] == 'Yes']['Country'].unique()
print("Countries with Restaurants Offering Online Delivery:\n", online_delivery_countries)
```

```python
locality_with_most_restaurants = df['Locality'].value_counts().idxmax()
print(f"Locality with the Most Restaurants: {locality_with_most_restaurants}")
```

```python
good_cuisine_restaurants = df[(df['Rating text'] == 'Good') & (df['Aggregate rating'] >= 3.0)]
print("Restaurants with Good Cuisine and Average Rating:\n", good_cuisine_restaurants[['Restaurant Name', 'Cuisines', 'Aggregate rating']])
```
